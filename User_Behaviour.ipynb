```python
 def plot_scatter_graph(user_behaviour):
    
    """
    plotting a scatter graph to show how screen time affect battery drain
    """
    
    plt.figure(dpi = 144, figsize = (5,5))
    
    plt.scatter(user_behaviour['Battery Drain (mAh/day)'], user_behaviour['Screen On Time (hours/day)'], color = 'blue', alpha = 0.3)
    plt.xlabel('Battery Drain (mAh/day)')
    plt.ylabel('Screen On Time (hrs/day)')
    plt.title('Graph of Screen Time to Battery Drain')
    return   
```

```python
plot_scatter_graph(user_behaviour)
```

```python
def plot_graph(user_behaviour, labels = ('App Usage Time', 'Data Usage', 'Gender')):
    
    """
    Plots histogram for number of App Usuage Time, Data usage and Gender on a single plot
    """
    
    fig, axs = plt.subplots(1, 3, dpi = 144, figsize = (10, 10))
    axs = axs.flatten()
     
    axs[0].hist(user_behaviour['App Usage Time (min/day)'], bins = 10, color = 'blue')
    axs[1].hist(user_behaviour['Data Usage (MB/day)'], bins = 10, color = 'green')
    axs[2].hist(user_behaviour['Gender'], bins = 10, color = 'brown' )
     
    for i, ax in enumerate(axs):
        ax.set_ylabel('Frequency', fontsize = 15)
        ax.set_xlabel(labels[i])
        
    fig.subplots_adjust(wspace = 0.5)
    return
```

```python
plot_graph(user_behaviour)
```

```python
def plot_user_behaviour_boxplot(user_behaviour):
    
    """
    Creates a box plot of age and number of apps installed  
    """
    plt.figure(dpi = 144, figsize= (5, 5))
    
    plt.boxplot([user_behaviour['Age'], user_behaviour['Number of Apps Installed']], labels = ('Age', 'Number of Apps Installed'), showmeans = True)
    plt.ylabel('Range')
    
    return
```

```python
plot_user_behaviour_boxplot(user_behaviour)
```
