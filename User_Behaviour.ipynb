```python
def plot_scatter_graph(user_behaviour):
    
    """
    plotting a scatter graph to show how screen time affect battery drain
    """
    
    plt.figure(dpi = 144)
    
    plt.scatter(user_behaviour['Battery Drain (mAh/day)'], user_behaviour['Screen On Time (hours/day)'], color = 'black')
    plt.xlabel('Battery Drain (mAh/day)')
    plt.ylabel('Screen Time (hrs/day)')
    plt.title('Graph of Screen Time to Battery Drain')
    
    return   
```

```python
plot_scatter_graph(user_behaviour)
```

```python
def plot_graph(user_behaviour, labels = ('Number of Apps', 'Data Usage', 'Gender')):
    
    """
    Plotting number of apps, Data usage and Gender on a single plot
    """
    fig, axs = plt.subplots(1, 3, dpi = 144, figsize = (15, 15))
    axs = axs.flatten()
     
    axs[0].hist(user_behaviour['Number of Apps Installed'], bins = 20, color = 'blue')
    axs[1].hist(user_behaviour['Data Usage (MB/day)'], bins = 20, color = 'green')
    axs[2].hist(user_behaviour['Gender'], bins = 10, color = 'brown' )
     
    for i, ax in enumerate(axs):
        ax.set_ylabel('Frequency', fontsize = 15)
        ax.set_title(labels[i])
    fig.subplots_adjust(wspace = 0.5)
    
    return
```

```python
plot_graph(user_behaviour)
```

```python
def plot_user_violin(user_behaviour):
    
    """
    A violin plot of Screen on time and User Behaviour class 
    """
    plt.figure(dpi = 144)
    
    plt.boxplot([user_behaviour['Screen On Time (hours/day)'], user_behaviour['User Behavior Class']], labels = ('Screen on Time', 'User behaviour class'), showmeans = True)
    plt.ylabel('Range')
    
    return
```

```python
plot_user_violin(user_behaviour)
```
